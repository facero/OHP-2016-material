## Interactive imaging from event files

Event files are tables listing the for each event the position, the energy, the time of arrival, etc
Therefore creating images (or spectra, or lightcurves) from the event file is just creating a 1D or 2D histogram.

Here we will explore how to create images without using the XMM SAS tools. This is a preliminary fun step to play interactively with your data by creating on-the fly images in any energy band.

Here I have specifically chosen a region of the sky where several energy components are observed and where playing with energy bands reveals a completely different view of the region showcasing the true power of spectro-imaging instruments.

```python
%matplotlib inline
from __future__ import  division, print_function
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, fixed
from astropy.io import fits
from astropy import table, wcs
from astropy.convolution import convolve,convolve_fft, Gaussian2DKernel
from matplotlib.colors import PowerNorm,LogNorm

#Small library containing a way to transform header information from XMM event list to 
#a WCS header in astropy
from  imaging_lib import hdrfromevt, ds9box, dist_circle



obslist=['0605960101']

#insts=['PN']
insts=['M1','M2']

RA  = np.array([])
DEC = np.array([])
E   = np.array([])


mydir='data'
obs=obslist[0]

for inst in insts:
#inst=insts[0]
    evtfile = '%s/%s/filtered%s.fits.gz'%(mydir,obs,inst)

    print(evtfile)
    #hdus = fits.open(evtfile)
    hdr = fits.getheader(evtfile, extname='EVENTS')
    #data = hdus['EVENTS'].data

    evt = table.Table.read( evtfile, hdu=1 )


    Xevt = evt['X']
    Yevt = evt['Y']
    E_ = evt['PI']

    #Getting keywords for column X (col6)and column Y (col7)
    hdr2 = hdrfromevt(hdr,'X', 'Y',6,7) 
    w = wcs.WCS(hdr2)
    RA_, DEC_ = w.wcs_pix2world(Xevt, Yevt, 0)

    RA = np.append(RA, RA_)
    DEC = np.append(DEC, DEC_)
    E = np.append(E, E_)
    print('%s events loaded '%(len(RA_)) )


print('Done')   
```

```python
def plot_image(RA,DEC,E,Emin=0.5,Emax=1,sigma=2):

#    print('\n%s events kept in total'%(len(RA)) )

    myE, = np.where((E > Emin*1e3) & (E < Emax*1e3))
    img, xedges, yedges = np.histogram2d(RA[myE], DEC[myE], bins=500,normed=False)

    E = E/1e3
    spectrum,energy = np.histogram(E ,bins = 10**np.linspace(np.log10(np.min(E)), np.log10(np.max(E)), 100))

    
    fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(16,7))
    gauss = Gaussian2DKernel(stddev=sigma)
    img_sm = convolve(img, gauss, boundary='extend')
    myextent=(RA.min(), RA.max(), DEC.min(), DEC.max())
    p1=ax1.imshow(img_sm.T,origin='lower',extent=myextent,aspect=1.5,norm=PowerNorm(0.5))
    ax1.ticklabel_format(useOffset=False)
    ax1.invert_xaxis()
    cbar=plt.colorbar(p1,fraction=0.045,ax=ax1)
    cbar.ax.set_ylabel('Counts/pixel')
    ax1.set_xlabel('RA')
    ax1.set_ylabel('DEC')
    
    ax2.errorbar(energy[:-1],spectrum,yerr=np.sqrt(spectrum),marker='.',linestyle='None',  capsize=0)
    ax2.axvline(x=Emin,ls='--',color='r')
    ax2.axvline(x=Emax,ls='--',color='r')
    ax2.set_xlim(np.min(E),np.max(E))
    ax2.set_ylim(5e2,3e4)
    ax2.loglog()
    ax2.set_xlabel('Energy (keV)')
    ax2.set_ylabel('Counts')
    ax2.set_aspect(0.8)
    plt.show()
    
interact(plot_image,RA=fixed(RA),DEC=fixed(DEC),E=fixed(E),
         Emin=(0.5,10,0.5),Emax=(1,13,0.5),sigma=(0.5,3,0.5)) 
```

Now we want to test a hardness ratio on that image to highlight the regions with high sprectal variations.<br/>
For this we can create a hardness map defined as :

$$HR = \left( \frac{H-S}{H+S} \right) $$

where S is the soft band (defined between E1 and E2) and H the hard band (E2=> E3). 


```python
def plot_HR(RA,DEC,E,E1=0.5,E2=1,E3=6,sigma=2):

    imsize=200
    myE, = np.where((E > E1*1e3) & (E < E2*1e3))
    S, xedges, yedges = np.histogram2d(RA[myE], DEC[myE], bins=imsize,normed=False)

    myE, = np.where((E > E2*1e3) & (E < E3*1e3))
    H, xedges, yedges = np.histogram2d(RA[myE], DEC[myE], bins=imsize,normed=False)

    HR = (H-S)/(H+S)
    dist = dist_circle(imsize,imsize,imsize/2.,imsize/2)
    HR[dist >75]=np.nan  #Hide border of CCD
    
    fig, ax = plt.subplots(1, 1, figsize=(7.5,7))
    gauss = Gaussian2DKernel(stddev=sigma)

    HR = convolve(HR, gauss, boundary='extend')

    myextent=(RA.min(), RA.max(), DEC.min(), DEC.max())
    p1=plt.imshow(HR.T,origin='lower',extent=myextent,aspect=1.5,cmap='seismic')
    ax.ticklabel_format(useOffset=False)
    ax.invert_xaxis()
    cbar=plt.colorbar(p1,fraction=0.05)
    cbar.ax.set_ylabel('(H-S)/(H+S)')
    plt.xlabel('RA')
    plt.ylabel('DEC')
    plt.show()
    
interact(plot_HR,RA=fixed(RA),DEC=fixed(DEC),E=fixed(E),
         E1=fixed(0.5),E2=(1,5,0.5),E3=fixed(5), sigma=(0.5,3,0.5)) 
```
